```python
from sympy import * 
x = Symbol('x') 
y = Symbol('y') 
z = (x + y) - (2*x+y) 
print("value of z is :" + str(z))
ans1 = diff(x**3,x)
print("ans1:",str(ans1))
ans2 = integrate(x**2,x)
print("ans2:",str(ans2))
ans4 = limit(tan(x)/x, x, 0) 
print("limit is : ", ans4) 
ans5 = diff(sin(2*x),x,2)
print("ans:",ans4)

```
